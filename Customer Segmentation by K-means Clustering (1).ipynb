# Clustering the data:

```python
from sklearn.cluster import KMeans
```

```python
## collecting the data:
clustering_data=df.drop(['Age','Work Experience','Gender_Label','CustomerID','Gender','Profession','Family Size','Age Category','gender income category','gender score category'],axis=1)
```

```python
clustering_data
```

Now finding the optimum number of cluster by WCSS:

```python
WCSS=[]
for i in range(1,11):
    Kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    Kmeans.fit(clustering_data)
    
    WCSS.append(Kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),WCSS)
plt.xlabel("number of Clusters")
plt.ylabel("WCSS")
plt.title("the elbow point plot")
plt.show()
```

From the above plot, there are two elbow point so we take the minimum elbow point having cluster  of "3".
